```python
import numpy as np
import pandas as pd
#import ace_tools as tools

#読み込む .npz ファイルのパスを指定
file_path = 'static_calib_params_1-month_v3.3.3.npz'

# npz を読み込み
data = np.load(file_path)

# DataFrame 用に情報を整形
info = []
for key in data.files:
    arr = data[key]
    # 要素数が多いときは形状のみ記載、それ以外は中身も表示
    if arr.size > 10:
        val = f"array(shape={arr.shape})"
    else:
        val = arr.tolist()
    info.append((key, arr.shape, str(arr.dtype), val))

df = pd.DataFrame(info, columns=['parameter', 'shape', 'dtype', 'value'])
file_path2 = 'static_calib_params_1-year_v3.3.3.npz'

# npz を読み込み
data2 = np.load(file_path2)

# DataFrame 用に情報を整形
info2 = []
for key in data.files:
    arr = data[key]
    # 要素数が多いときは形状のみ記載、それ以外は中身も表示
    if arr.size > 10:
        val = f"array(shape={arr.shape})"
    else:
        val = arr.tolist()
    info2.append((key, arr.shape, str(arr.dtype), val))

df2 = pd.DataFrame(info2, columns=['parameter', 'shape', 'dtype', 'value'])

# merge df and df2
df = pd.merge(df, df2, on='parameter', suffixes=('_1-month', '_1-year'))
df

# ユーザーに表示
#tools.display_dataframe_to_user('Static Calibration Parameters', df)
```
