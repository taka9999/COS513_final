```python
import pandas as pd
import matplotlib.pyplot as plt

# ----- ユーザー設定 -----
csv_file   = "DGP_IR_scenarios_1-year_v3.1.csv"  # 読み込むCSVファイル名
target_date = "2023-02-28"                     # プロット対象の日付

# 1. CSV読み込み
#    ・date 列を日時型に  
df = pd.read_csv(csv_file, parse_dates=["date"])

# 2. 日付フィルタ
df_sel = df[df["date"] == pd.to_datetime(target_date)]

if df_sel.empty:
    raise ValueError(f"No data found for date {target_date}")

# 3. シナリオ列だけ抜き出し（1–50年）
#    CSVのヘッダーが文字列 "1","2",… の場合：
maturities = list(range(1, 51))
col_names   = [str(m) for m in maturities]
data = df_sel[col_names].values  # shape=(n_sim, 50)

# 4. プロット
plt.figure(figsize=(10, 6))
for curve in data:
    plt.plot(maturities, curve, alpha=0.05)
# 平均カーブ
mean_curve = data.mean(axis=0)
plt.plot(maturities, mean_curve, linewidth=2, label="Mean Scenario")

plt.xlabel("Maturity (years)")
plt.ylabel("Yield")
plt.title(f"Interest Rate Scenarios on {target_date}")
plt.legend()
plt.grid(True)
plt.show()
```
